```python
# Задание № 2 
import numpy as np
t = np.linspace (0, 10, 51)
print(t)
f = np.cos(t)
print(f)
```

```python
title_dict={'fontsize': 14, 'fontweight': 'bold', 'color': 'Grey'}
lable_dict={'fontsize': 10, 'fontweight': 'bold', 'color': 'Grey'}

plt.plot(t, f, color='green')
plt.title('график f(t)', fontdict=title_dict, loc='left')
plt.xlabel('Значения t', fontdict=lable_dict)
plt.ylabel('Значения f', fontdict=lable_dict)
plt.axis([0.5, 9.5, -2.5, 2.5])

plt.show()
```

```python
# Задание № 3 
import matplotlib.pyplot as plt
x = np.linspace (-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

title_dict={'fontsize': 12, 'fontweight': 'bold', 'color': 'Gold'}

fig, ax = plt.subplots(nrows=2, ncols=2) # 2 столбца 2 строки.
ax1, ax2, ax3, ax4 = ax.flatten() # "упакуем" фигуры axes в figure
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
ax1.set_title('График y1', fontdict=title_dict)
ax2.set_title('График y2', fontdict=title_dict)
ax3.set_title('График y3', fontdict=title_dict)
ax4.set_title('График y4', fontdict=title_dict)
ax1.set_xlim([-5, 5]) # ставим лимит на график в левом верхнем углу 
fig.set_size_inches(8, 6) 
plt.subplots_adjust(wspace=0.3, hspace=0.3) # ставим вертикальный и горизонтальный зазор. 
plt.show()
```
